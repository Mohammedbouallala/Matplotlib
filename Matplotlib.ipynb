```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # needed for 3D plotting
plt.rcParams.update({'font.size': 10})
```

```python
x1 = np.linspace(-10, 10, 100)
y1 = 2 * x1**3 - 5 * x1**2 + 3 * x1 - 7

plt.figure(figsize=(10, 6))
plt.plot(x1, y1, 'b-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Polynomial Plot: y = 2x³ - 5x² + 3x - 7")
plt.show()
```

```python
x2 = np.linspace(0.1, 10, 500)
y_exp = np.exp(x2)
y_log = np.log(x2)

plt.figure(figsize=(10, 6))
plt.plot(x2, y_exp, 'r-', label='exp(x)')
plt.plot(x2, y_log, 'g--', label='log(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Exponential and Logarithmic Functions")
plt.grid(True)
plt.legend()
plt.savefig("exponential_logarithmic.png", dpi=100)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# First subplot: tan and arctan (limit tan to avoid extreme values)
x3a = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y_tan = np.tan(x3a)
y_arctan = np.arctan(x3a)
axs[0].plot(x3a, y_tan, label='tan(x)')
axs[0].plot(x3a, y_arctan, label='arctan(x)')
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].set_title("tan(x) and arctan(x)")
axs[0].legend()
axs[0].set_ylim(-10, 10)  # restrict tan(x) range

# Second subplot: sinh and cosh
x3b = np.linspace(-2, 2, 400)
y_sinh = np.sinh(x3b)
y_cosh = np.cosh(x3b)
axs[1].plot(x3b, y_sinh, label='sinh(x)')
axs[1].plot(x3b, y_cosh, label='cosh(x)')
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].set_title("sinh(x) and cosh(x)")
axs[1].legend()

plt.tight_layout()
plt.show()

# Histogram of 1000 values from a normal distribution
n = np.random.normal(size=1000)
plt.figure(figsize=(10, 6))
plt.hist(n, bins=30, color='purple', edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlim(n.min(), n.max())
plt.show()
```

```python
x5 = np.linspace(0, 10, 500)
noise = np.random.normal(scale=0.3, size=500)
y5 = np.sin(x5) + noise

# Vary marker sizes and colors based on y5 values.
sizes = 20 + 100 * (np.abs(y5) / np.max(np.abs(y5)))
colors = y5

plt.figure(figsize=(10, 6))
plt.scatter(x5, y5, s=sizes, c=colors, cmap='viridis', alpha=0.7)
plt.grid(True)
plt.xticks([])  # hide x ticks
plt.yticks([])  # hide y ticks
plt.title("Scatter Plot with Varying Marker Sizes and Colors")
plt.savefig("scatter_plot.pdf")
plt.show()
```

```python
x6 = np.linspace(-5, 5, 200)
y6 = np.linspace(-5, 5, 200)
X6, Y6 = np.meshgrid(x6, y6)
Z6 = np.sin(np.sqrt(X6**2 + Y6**2))

plt.figure(figsize=(10, 6))
contours = plt.contour(X6, Y6, Z6, cmap='plasma')
plt.clabel(contours, inline=True, fontsize=8)
plt.title("Contour Plot of f(x, y) = sin(sqrt(x²+y²))")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x7 = np.arange(-5, 5.25, 0.25)
y7 = np.arange(-5, 5.25, 0.25)
X7, Y7 = np.meshgrid(x7, y7)
Z7 = np.cos(np.sqrt(X7**2 + Y7**2))

fig = plt.figure(figsize=(10, 8))
ax7 = fig.add_subplot(111, projection='3d')
surf = ax7.plot_surface(X7, Y7, Z7, cmap='coolwarm', edgecolor='none')
ax7.view_init(elev=30, azim=45)
ax7.set_title("3D Surface Plot of Z = cos(sqrt(x²+y²))")
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

```python
x8 = np.linspace(-2, 2, 10)
y8_1 = x8**2
y8_2 = x8**3
y8_3 = x8**4

plt.figure(figsize=(10, 6))
plt.plot(x8, y8_1, 'ro-', label='x²')
plt.plot(x8, y8_2, 'bs--', label='x³')
plt.plot(x8, y8_3, 'g^:', label='x⁴')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Different Line and Marker Styles")
plt.legend()
plt.show()
```

```python
x9 = np.linspace(1, 100, 50)
y9_1 = 2 ** x9
y9_2 = np.log2(x9)

plt.figure(figsize=(12, 6))
plt.plot(x9, y9_1, 'm-', label='2^x')
plt.plot(x9, y9_2, 'c--', label='log₂(x)')
plt.yscale('log')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Functions with Logarithmic y-scale")
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))
ax10_1 = fig.add_subplot(121, projection='3d')
ax10_2 = fig.add_subplot(122, projection='3d')

# Use the same surface data (X7, Y7, Z7)
surf1 = ax10_1.plot_surface(X7, Y7, Z7, cmap='coolwarm', edgecolor='none')
ax10_1.view_init(elev=30, azim=45)
ax10_1.set_title("Viewing Angle: elev=30, azim=45")
fig.colorbar(surf1, ax=ax10_1, shrink=0.5)

surf2 = ax10_2.plot_surface(X7, Y7, Z7, cmap='coolwarm', edgecolor='none')
ax10_2.view_init(elev=60, azim=120)
ax10_2.set_title("Viewing Angle: elev=60, azim=120")
fig.colorbar(surf2, ax=ax10_2, shrink=0.5)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax11 = fig.add_subplot(111, projection='3d')
# Create grid for wireframe
x11 = np.arange(-5, 5.25, 0.25)
y11 = np.arange(-5, 5.25, 0.25)
X11, Y11 = np.meshgrid(x11, y11)
Z11 = np.sin(X11) * np.cos(Y11)
ax11.plot_wireframe(X11, Y11, Z11, color='green')
ax11.view_init(elev=20, azim=30)
ax11.set_title("3D Wireframe Plot of Z = sin(x) * cos(y)")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax12 = fig.add_subplot(111, projection='3d')
x12 = np.arange(-5, 5.25, 0.25)
y12 = np.arange(-5, 5.25, 0.25)
X12, Y12 = np.meshgrid(x12, y12)
Z12 = np.exp(-0.1 * (X12**2 + Y12**2))
# Plot 3D contours
contour3d = ax12.contour3D(X12, Y12, Z12, 50, cmap='viridis')
ax12.set_xlabel("X")
ax12.set_ylabel("Y")
ax12.set_zlabel("Z")
ax12.set_title("3D Contour Plot of Z = exp(-0.1*(X²+Y²))")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax13 = fig.add_subplot(111, projection='3d')
t13 = np.linspace(0, 2 * np.pi, 100)
X13 = np.sin(t13)
Y13 = np.cos(t13)
Z13 = t13
ax13.plot3D(X13, Y13, Z13, color='magenta')
ax13.set_xlabel("X")
ax13.set_ylabel("Y")
ax13.set_zlabel("Z")
ax13.set_title("3D Parametric Plot")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax14 = fig.add_subplot(111, projection='3d')
x14 = np.linspace(-5, 5, 10)
y14 = np.linspace(-5, 5, 10)
X14, Y14 = np.meshgrid(x14, y14)
Z14 = np.exp(-0.1 * (X14**2 + Y14**2))
# Dimensions for bars:
dx = (x14[1] - x14[0]) * 0.8
dy = (y14[1] - y14[0]) * 0.8
xpos = X14.ravel() - dx/2
ypos = Y14.ravel() - dy/2
dz = Z14.ravel()

bars = ax14.bar3d(xpos, ypos, np.zeros_like(dz), dx, dy, dz, shade=True, color='cyan')
ax14.set_xlabel("X")
ax14.set_ylabel("Y")
ax14.set_zlabel("Z")
ax14.set_title("3D Bar Plot")
ax14.view_init(elev=30, azim=45)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax15 = fig.add_subplot(111, projection='3d')
vec_range = np.linspace(-5, 5, 10)
X15, Y15, Z15 = np.meshgrid(vec_range, vec_range, vec_range)
U15 = -Y15
V15 = X15
W15 = Z15
ax15.quiver(X15, Y15, Z15, U15, V15, W15, length=0.5, normalize=True, color='red')
ax15.set_title("3D Vector Field")
ax15.set_xlabel("X")
ax15.set_ylabel("Y")
ax15.set_zlabel("Z")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax18 = fig.add_subplot(111, projection='3d')
x18 = np.arange(-5, 5.1, 0.1)
y18 = np.arange(-5, 5.1, 0.1)
X18, Y18 = np.meshgrid(x18, y18)
Z18 = np.sin(np.sqrt(X18**2 + Y18**2))
# Using contourf with projection (note: this creates a filled contour on a plane)
contf = ax18.contourf(X18, Y18, Z18, zdir='z', offset=-1.5, cmap='viridis')
ax18.set_xlabel("X")
ax18.set_ylabel("Y")
ax18.set_zlabel("Z")
ax18.set_title("3D Filled Contour Plot")
fig.colorbar(contf, ax=ax18, shrink=0.5)
plt.show()
```
